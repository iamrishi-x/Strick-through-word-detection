```python

# Plot Accuracy
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, train_accuracy, label="Train Accuracy", marker="o")
plt.plot(epochs, val_accuracy, label="Validation Accuracy", marker="s")
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.title("Training vs Validation Accuracy Over Epochs")
plt.legend()
plt.grid(True)

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(epochs, train_loss, label="Train Loss", marker="o", color="red")
plt.plot(epochs, val_loss, label="Validation Loss", marker="s", color="purple")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss Over Epochs")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Simulated number of epochs
num_epochs = 20

# Initialize loss and accuracy lists
losses = []
accuracies = []

# Set up the live plot
plt.ion()
fig, ax = plt.subplots(2, 1, figsize=(8, 6))

for epoch in range(1, num_epochs + 1):
    # Simulating loss decreasing and accuracy increasing
    loss = np.exp(-epoch / 5) + np.random.uniform(0, 0.05)
    accuracy = 1 - loss + np.random.uniform(0, 0.02)
    
    losses.append(loss)
    accuracies.append(accuracy)
    
    # Clear previous plots
    ax[0].cla()
    ax[1].cla()
    
    # Plot loss
    ax[0].plot(range(1, epoch + 1), losses, marker='o', linestyle='-', color='r')
    ax[0].set_title('Training Loss Over Epochs')
    ax[0].set_xlabel('Epoch')
    ax[0].set_ylabel('Loss')
    ax[0].grid(True)
    
    # Plot accuracy
    ax[1].plot(range(1, epoch + 1), accuracies, marker='o', linestyle='-', color='g')
    ax[1].set_title('Training Accuracy Over Epochs')
    ax[1].set_xlabel('Epoch')
    ax[1].set_ylabel('Accuracy')
    ax[1].grid(True)
    
    # Draw the updated plot
    plt.pause(0.5)
    
    print(f"Epoch {epoch}/{num_epochs} - Loss: {loss:.4f} - Accuracy: {accuracy:.4f}")
    time.sleep(0.3)

plt.ioff()
plt.show()
```
